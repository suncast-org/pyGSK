# SK Test — programmatic example

Demonstrates how to call `pygsk.runtests.run_sk_test()` directly in Python,
inspect its results, and visualize the SK histogram using
`pygsk.plot.plot_sk_histogram()`.

```python
# --- Imports ---
from pygsk.runtests import run_sk_test
from pygsk.plot import plot_sk_histogram
import os
```

```python
# --- Parameters ---
M, N, d = 128, 64, 1.0
PFA = 1e-3
NS, NF = 10000, 4
MODE = "burst"
FIGDIR = "_figs"
SHOW = True
SAVE_PATH = f"{FIGDIR}/example_sk_dual_hist.png"

os.makedirs(FIGDIR, exist_ok=True)
```

```python
# --- Run canonical SK validation ---
result = run_sk_test(
    M=M, N=N, d=d, pfa=PFA,
    ns=NS, nf=NF, mode=MODE,
    verbose=True, plot=False
)
print("Returned keys:", list(result.keys()))
print(f"Empirical PFA (two-sided): {result['pfa_empirical']:.3e}")
print(f"Expected  PFA (two-sided): {result['pfa_expected']:.3e}")
```

```python
# --- Visualize ---
plot_sk_histogram(
    result,
    show=SHOW,
    save_path=SAVE_PATH,
    no_context=False,
)
print("Saved", SAVE_PATH)
```
